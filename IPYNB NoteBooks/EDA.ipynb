```python
# From our other notebook we created 5 csv files.
# we will individually load and analyze each of these files

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import os
import sys
```

```python
strength_tincture_market = pd.read_csv("..\\data\\strength_tincture_market.csv")
mind_tincture_market = pd.read_csv("..\\data\\mind_tincture_market.csv")
vitality_tincture_market = pd.read_csv("..\\data\\vitality_tincture_market.csv")
dexterity_tincture_market = pd.read_csv("..\\data\\dexterity_tincture_market.csv")
intelligence_tincture_market = pd.read_csv("..\\data\\intelligence_tincture_market.csv")
```

```python
strength_tincture_market.head()
```

```python
# Break down the timestamp into year, month, day, hour, minute, second, day of the week, and week of the year
strength_tincture_market['timestamp'] = pd.to_datetime(strength_tincture_market['timestamp'])
strength_tincture_market['year'] = strength_tincture_market['timestamp'].dt.year
strength_tincture_market['month'] = strength_tincture_market['timestamp'].dt.month
strength_tincture_market['day'] = strength_tincture_market['timestamp'].dt.day
strength_tincture_market['hour'] = strength_tincture_market['timestamp'].dt.hour
strength_tincture_market['minute'] = strength_tincture_market['timestamp'].dt.minute
strength_tincture_market['second'] = strength_tincture_market['timestamp'].dt.second
strength_tincture_market['day_of_week'] = strength_tincture_market['timestamp'].dt.dayofweek
strength_tincture_market['day_of_week'] = strength_tincture_market['day_of_week'].map({0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'})
strength_tincture_market['week_of_year'] = strength_tincture_market['timestamp'].dt.isocalendar().week
# save the data to a new csv file
strength_tincture_market.to_csv("..\\data\\strength_tincture_market.csv", index=False)

strength_tincture_market.head()
```
